```python
import schemdraw
import schemdraw.elements as elm
import matplotlib
```

```python
with schemdraw.Drawing() as d:
    d += (l1 := elm.Line(label='A'))
    # d += elm.Dot()
    # d += elm.Line(label='$V_{in}$').dot(open=True).right().hold()
    d += elm.Resistor().down().length(1.75)
    d += elm.Dot()
    # d += elm.Line(label='$V_{out}$').dot(open=True).right().hold()
    d += elm.Resistor().down().length(1.75)
    d += (l2 := elm.Line().left())
    d += elm.Ground()
    d += (PS := elm.SourceV(label='V').endpoints(l2.end, l1.start))
d.save('DC2-1.png')
```

```python
with schemdraw.Drawing() as d:
    d += (l1 := elm.Line(label='B'))
    #d += elm.Dot()
    # d += elm.Line(label='$V_{in}$').dot(open=True).right().hold()
    d += (R1 := elm.Resistor().down())
    # d += elm.Line(label='$V_{out}$').dot(open=True).right().hold()
    #d += elm.Line().down()
    d += (l2 := elm.Line().left())
    d += elm.Ground()
    d += (PS := elm.SourceV(label='V').endpoints(l2.end, l1.start))
d.save('DC2-2.png')
```

```python
with schemdraw.Drawing() as d:
    d += (l1 := elm.Line(label='C').length(2))
    # d += elm.Line(label='$V_{in}$').dot(open=True).right().hold()
    d += (R1 := elm.Resistor().down())
    #d += elm.Dot()
    # d += elm.Line(label='$V_{out}$').dot(open=True).right().hold()
    #d += elm.Line().down()
    d += (l2 := elm.Line().left().length(2))
    d += elm.Ground()
    d += (PS := elm.SourceV(label='V').endpoints(l2.end, l1.start))
    d += elm.Line().at(l1.end).right().length(1)
    d += elm.Resistor().down()
    #d += elm.Line().down()
    d += elm.Line().left().length(1)
    
d.save('DC2-3.png')
```

```python
with schemdraw.Drawing() as d:
    d += (l1 := elm.Line())
    # d += elm.Dot()
    # d += elm.Line(label='$V_{in}$').dot(open=True).right().hold()
    d += (R1 := elm.Resistor(label='V1', ofst=(0, -1)).down())
    d += elm.Dot()
    # d += elm.Line(label='$V_{out}$').dot(open=True).right().hold()
    d += (R2 := elm.Resistor(label='V2', ofst=(0, -1)).down())
    d += (l2 := elm.Line().left())
    d += elm.Ground()
    d += (PS := elm.SourceV(label='V').endpoints(l2.end, l1.start))
    d += elm.CurrentLabel().at(R1).label('R1', ofst=(0, -1.5))
    d += elm.CurrentLabel().at(R2).label('R2', ofst=(0, -1.5))
d.save('DC2-5.png')
```

```python
with schemdraw.Drawing() as d:
    d += (l1 := elm.Line(label='A'))
    d += elm.Dot()
    # d += elm.Line(label='$V_{in}$').dot(open=True).right().hold()
    d += elm.Resistor().down().length(2)
    d += elm.Dot()
    # d += elm.Line(label='$V_{out}$').dot(open=True).right().hold()
    d += elm.Resistor().down().length(4)
    d += elm.Dot()
    d += (l2 := elm.Line().left())
    d += elm.Ground()
    d += (PS := elm.SourceV(label='5V').endpoints(l2.end, l1.start))
    
    d += elm.Line().at(l1.end).right().length(1)
    d += elm.Resistor().down().length(2)
    d += elm.Line().left().length(1)    
    
d.save('DC2-6.png')
```

```python
with schemdraw.Drawing() as d:
    d += (l1 := elm.Line(label='B'))
    d += (dot1 := elm.Dot())
    # d += elm.Line(label='$V_{in}$').dot(open=True).right().hold()
    d += elm.Resistor().down().length(2)
    d += (dot2 := elm.Dot())
    # d += elm.Line(label='$V_{out}$').dot(open=True).right().hold()
    d += elm.Resistor().down().length(2)
    d += (dot3 := elm.Dot())
    d += elm.Resistor().down().length(2)
    d += elm.Dot()
    d += (l2 := elm.Line()).left()
    d += elm.Ground()
    d += (PS := elm.SourceV(label='5V').endpoints(l2.end, l1.start))
    
    d += elm.Line().at(dot1.end).right().length(1)
    d += elm.Resistor().down().length(2)
    d += elm.Line().left().length(1)    
    
d.save('DC2-7.png')
```
